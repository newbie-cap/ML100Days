```python
# 做完特徵工程前的所有準備 (與前範例相同)
import pandas as pd
import numpy as np
import copy
from sklearn.preprocessing import LabelEncoder, MinMaxScaler
from sklearn.model_selection import cross_val_score
from sklearn.linear_model import LogisticRegression
import warnings
warnings.filterwarnings('ignore')

data_path = 'D:/Data/'
df = pd.read_csv(data_path + 'titanic_train.csv')
df1 = df #new
train_Y = df['Survived']
df = df.drop(['PassengerId', 'Survived'] , axis=1)
df.head(10)
```

```python
df.dtypes
```

```python
df.shape
```

```python
# 想法:依傳票等級分類，各自做群聚編碼
df_Pclass1 = []
df_Pclass2 = []
df_Pclass3 = []

i = 0 
for i in range(len(df)):
    if df['Pclass'][i] == 1 :
        df_Pclass1.append(df.iloc[i,:])
    elif df['Pclass'][i] == 2 :
        df_Pclass2.append(df.iloc[i,:])
    elif df['Pclass'][i] == 3 :        
        df_Pclass3.append(df.iloc[i,:])
    else:
        print(i)
    i+=1     
    
df_Pclass1 = pd.DataFrame(df_Pclass1)
df_Pclass2 = pd.DataFrame(df_Pclass2)
df_Pclass3 = pd.DataFrame(df_Pclass3)

print(f' df_Pclass1 : {len(df_Pclass1)} \n')
print(f' df_Pclass2 : {len(df_Pclass2)} \n')
print(f' df_Pclass3 : {len(df_Pclass3)} \n')

df_Pclass1.head()
```

```python
# 取一個類別型欄位, 與一個數值型欄位, 做群聚編碼
"""
Your Code Here
"""
# df_Pclass1
df_Pclass1['Embarked'] = df_Pclass1['Embarked'].fillna('None')
mean_df_Pclass1 = df_Pclass1.groupby(['Embarked'])['Fare'].mean().reset_index()
mode_df_Pclass1 = df_Pclass1.groupby(['Embarked'])['Fare'].apply(lambda x: x.mode()[0]).reset_index()
median_df_Pclass1 = df_Pclass1.groupby(['Embarked'])['Fare'].median().reset_index()
max_df_Pclass1 = df_Pclass1.groupby(['Embarked'])['Fare'].max().reset_index()

temp_Pclass1 = pd.merge(mean_df_Pclass1, mode_df_Pclass1, how='left', on=['Embarked'])
temp_Pclass1 = pd.merge(temp_Pclass1, median_df_Pclass1, how='left', on=['Embarked'])
temp_Pclass1 = pd.merge(temp_Pclass1, max_df_Pclass1, how='left', on=['Embarked'])
temp_Pclass1.columns = ['Embarked', 'Embarked_Fare_Mean', 'Embarked_Fare_Mode', 'Embarked_Fare_Median', 'Embarked_Fare_Max']
temp_Pclass1
```

```python
Pclass1 = pd.merge(df_Pclass1, temp_Pclass1, how='left', on=['Embarked'])
Pclass1 = Pclass1.drop(['Embarked'], axis=1)
Pclass1.head()
```

```python
# 取一個類別型欄位, 與一個數值型欄位, 做群聚編碼
# df_Pclass2
df_Pclass2['Embarked'] = df_Pclass2['Embarked'].fillna('None')
mean_df_Pclass2 = df_Pclass2.groupby(['Embarked'])['Fare'].mean().reset_index()
mode_df_Pclass2 = df_Pclass2.groupby(['Embarked'])['Fare'].apply(lambda x: x.mode()[0]).reset_index()
median_df_Pclass2 = df_Pclass2.groupby(['Embarked'])['Fare'].median().reset_index()
max_df_Pclass2 = df_Pclass2.groupby(['Embarked'])['Fare'].max().reset_index()

temp_Pclass2 = pd.merge(mean_df_Pclass2, mode_df_Pclass2, how='left', on=['Embarked'])
temp_Pclass2 = pd.merge(temp_Pclass2, median_df_Pclass2, how='left', on=['Embarked'])
temp_Pclass2 = pd.merge(temp_Pclass2, max_df_Pclass2, how='left', on=['Embarked'])
temp_Pclass2.columns = ['Embarked', 'Embarked_Fare_Mean', 'Embarked_Fare_Mode', 'Embarked_Fare_Median', 'Embarked_Fare_Max']
temp_Pclass2
```

```python
Pclass2 = pd.merge(df_Pclass2, temp_Pclass2, how='left', on=['Embarked'])
Pclass2 = Pclass2.drop(['Embarked'], axis=1)
Pclass2.head()
```

```python
# 取一個類別型欄位, 與一個數值型欄位, 做群聚編碼
# df_Pclass3
df_Pclass3['Embarked'] = df_Pclass3['Embarked'].fillna('None')
mean_df_Pclass3 = df_Pclass3.groupby(['Embarked'])['Fare'].mean().reset_index()
mode_df_Pclass3 = df_Pclass3.groupby(['Embarked'])['Fare'].apply(lambda x: x.mode()[0]).reset_index()
median_df_Pclass3 = df_Pclass3.groupby(['Embarked'])['Fare'].median().reset_index()
max_df_Pclass3 = df_Pclass3.groupby(['Embarked'])['Fare'].max().reset_index()

temp_Pclass3 = pd.merge(mean_df_Pclass3, mode_df_Pclass3, how='left', on=['Embarked'])
temp_Pclass3 = pd.merge(temp_Pclass3, median_df_Pclass3, how='left', on=['Embarked'])
temp_Pclass3 = pd.merge(temp_Pclass3, max_df_Pclass3, how='left', on=['Embarked'])
temp_Pclass3.columns = ['Embarked', 'Embarked_Fare_Mean', 'Embarked_Fare_Mode', 'Embarked_Fare_Median', 'Embarked_Fare_Max']
temp_Pclass3
```

```python
Pclass3 = pd.merge(df_Pclass3, temp_Pclass3, how='left', on=['Embarked'])
Pclass3 = Pclass3.drop(['Embarked'], axis=1)
Pclass3.tail()
```

```python
# 合併Pclass1, Pclass2, Pclass3表格
Pclass = pd.concat([Pclass1,Pclass2,Pclass3],axis=0, ignore_index=True)
Pclass
```

```python
#只取 int64, float64 兩種數值型欄位, 存於 num_features 中
num_features = []
for dtype, feature in zip(Pclass.dtypes, Pclass.columns):
    if dtype == 'float64' or dtype == 'int64':
        num_features.append(feature)
print(f'{len(num_features)} Numeric Features : {num_features}\n')

# 削減文字型欄位, 只剩數值型欄位
Pclass = Pclass[num_features]
Pclass = Pclass.fillna(-1)
MMEncoder = MinMaxScaler()
print(Pclass.head())
print(Pclass.tail())
```

```python
### 原始特徵(兩種數值型欄位)
Pclass_minus = Pclass.drop(['Embarked_Fare_Mean', 'Embarked_Fare_Mode', 'Embarked_Fare_Median', 'Embarked_Fare_Max'], axis=1)
Pclass_minus
```
